```python
#Questão 2 - Importando o Dataset

import pandas as pd
df = pd.read_csv('https://raw.githubusercontent.com/wagnernsct/NF3-BIG-DATA-WAGNER/master/cursos-prouni.csv', sep = ',')
type(df)
df


```

```python
# Questão 2 - Letra A | Efetuar a limpeza das colunas de notas: onde tiver NaN (Not a Number), substituir por 0,0.
colunas_notas = df.filter(like='nota')
colunas_notas = colunas_notas.fillna(0.0)
df[colunas_notas.columns] = colunas_notas
print(df)
```

```python
# Questão 2 - letra B | Agrupe os dados pelo grau (Bacharelado, Licenciatura, etc).

grau = df.groupby("grau")

for x, grupo in grau:
    print("Grupo:", x)
    print(grupo.describe())
    print("\n")
```

```python
# Questão 2 - letra C | Agrupe os dados pelos cursos de Matemática, Medicina e Pedagogia.

curso_busca = ['Matemática', 'Medicina', 'Pedagogia']
df_curso_busca = df[df['curso_busca'].isin(curso_busca)]

grupo_curso = df_curso_busca.groupby('curso_busca')

matematica = grupo_curso.get_group('Matemática')
medicina = grupo_curso.get_group('Medicina')
pedagogia = grupo_curso.get_group('Pedagogia')


print("Matemática:")
print(matematica)
print("\n")


print("Medicina:")
print(medicina)
print("\n")

print("Pedagogia:")
print(pedagogia)
print("\n")
```

```python
# Questão 2 - letra D | Agrupe os dados por Estado e obtenha a média de notas de corte por Estado

df.groupby("uf_busca")["nota_integral_ampla", "nota_integral_cotas", "nota_parcial_ampla", "nota_parcial_cotas"].mean()


```

```python
# Questão 2 - letra E | Agrupe os dados pelos cursos Tecnológicos.

curso = df.groupby("grau")
curso.get_group("Tecnológico").groupby("grau").describe()
```

```python
# Questão 2 - letra F | Elimine a coluna “cidade_filtro” do dataframe.

df.drop("cidade_filtro", axis = 1, inplace = True)




```

```python
# Questão 2 - letra F | Elimine a coluna “cidade_filtro” do dataframe. (Conferindo se a coluna foi excluída com êxito).
df.head()
```

```python
# Questão 2 - letra G | Apresente a média das mensalidades dos cursos de Medicina.

media_medicina = df.loc[df['curso_busca'] == 'Medicina', 'mensalidade'].mean()
print("Média das mensalidades dos cursos de Medicina:", media_medicina)


```

```python
# Questão 2 - Letra H |  Média das notas de corte dos cursos de tempo integral.

media_notas = df[(df['grau'] == 'Bacharelado') & (df['nota_integral_ampla'])]
notas = media_notas["nota_integral_ampla"].describe()
notas
```

```python
# Questão 2 - Letra J | Gráfico comparativo entre o grau dos cursos (Bacharelado, Licenciatura, Tecnologia, etc) pelas Notas Integral de Cotas.

grafico = df.groupby("grau")["nota_integral_cotas"].std()
grafico
grafico.plot.bar(x="grau", y = "nota_integral_ampla")
```
